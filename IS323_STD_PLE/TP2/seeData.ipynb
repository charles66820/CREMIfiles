```python
filteredData = data[["keywords"]]
filteredData.to_csv('out.csv', index=False)
```

```python
keywordsSet = {None}
wordCount = 0
for i in filteredData.iterrows():
    r = i[1][0]
    if type(r) is float:
        continue
    for w in r.split(";"):
        wordCount += 1
        keywordsSet.add(w.replace("\"", ""))
```

```python
print(wordCount)
len(keywordsSet)

# mapReduce keyword/decade : 168787 - 1 (header)

# all keywords with duplicate mapReduce: 1060964
# all keywords with duplicate python: 1060970 - 1 (header) - 5 ???
# avec mapReduce : 130365 - 1 (header)
# avec python : 130366 - 1 (header) - 1 (None)
```

```python
with open('oui2.csv', 'a') as f:
    for k in keywordsSet:
        if k != None:
            f.write(k + "\n")
```
