```python
import numpy as np
import pandas as pa
```

```python
dataset = pa.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv")
data = pa.DataFrame(dataset)
```

```python
euData = data[data["continent"] == "Europe"]
franceData = euData[euData["iso_code"] == "FRA"]
franceData.head()
```

```python
data.info()
```

```python
casesData = franceData[["date", "new_cases", "new_cases_smoothed", "location"]]
casesData
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

X = np.linspace(0, len(casesData), len(casesData))
Y = casesData["new_cases"]

plt.plot(X, casesData["new_cases"], c="red")
plt.plot(X, casesData["new_cases_smoothed"], c="blue")
```
